#### Datasets

First of all, let's create our dataset and data loaders. We will use training, validation and test sets. You will measure your best model based on its performance on the validation set. **REMEMBER**: your best test performance is not the highest test accuracy you achieve, but it's the test accuracy that you achieve at the highest validation accuracy!

In order to load our data, we will use the `ImageFolder` class. This class receives as argument the path to a directory which contains **one sub-directory per class**. For example, you can see that the `dog_images/train` directory in the Google Drive share has this structure. The class will return a `Dataset` instance, with the usual interface (`len` to get the number of samples, `[]` to access each sample), that you can pass to a `DataLoader` instance.

Also, note that you training transformation (with data augmentation) will be different from your test transformations (without data augmentation: for example, you don't need random cropping).

```python
# Imports
import torch
from torchvision.datasets import ImageFolder
import torchvision.transforms as T
from torch.utils.data import DataLoader

# Data loading options
batch_size = 64
load_size = 230
crop_size = 224
norm_mean = (0.5, 0.5, 0.5)
norm_std = (0.5, 0.5, 0.5)

# Train transform
train_transform = T.Compose([
        T.Resize(load_size),
        T.RandomCrop(crop_size),
        T.RandomHorizontalFlip(),
        T.ToTensor(),
        T.Normalize(norm_mean, norm_std)
])

# Test transform
test_transform = T.Compose([
        T.Resize(load_size),
        T.CenterCrop(crop_size),
        T.ToTensor(),
        T.Normalize(norm_mean, norm_std) 
])

# Instantiate datasets
dog_train_dataset = ImageFolder(os.path.join(root_dir, "dog_images", "train"), transform=train_transform)
dog_val_dataset = ImageFolder(os.path.join(root_dir, "dog_images", "valid"), transform=test_transform)
dog_test_dataset = ImageFolder(os.path.join(root_dir, "dog_images", "test"), transform=test_transform)

# Get number of classes (we'll need it in the model)
num_classes = len(dog_train_dataset.classes)

# Print dataset statistics
print(f"Num. classes: {num_classes}")
print(f"Num. train samples: {len(dog_train_dataset)}")
print(f"Num. valid. samples: {len(dog_val_dataset)}")
print(f"Num. test samples: {len(dog_test_dataset)}")


# Instantiate data loaders
train_loader = DataLoader(dog_train_dataset, batch_size=batch_size, num_workers=1, shuffle=True)
val_loader   = DataLoader(dog_val_dataset,   batch_size=batch_size, num_workers=1, shuffle=False)
test_loader  = DataLoader(dog_test_dataset,  batch_size=batch_size, num_workers=1, shuffle=False)

loaders = {"train": train_loader,
           "val": val_loader,
           "test": test_loader}
```

#### Model definition

Define your CNN architecture.

```python
# Imports
import torch.nn as nn
import torch.nn.functional as F

# Define model
class Model(nn.Module):
    #
    def __init__(self):
      super().__init__();

      self.conv_layers = nn.Sequential(
          nn.Conv2d(3, 8, kernel_size = 3, padding = 0, stride = 1),
          nn.ReLU(),
          nn.BatchNorm2d(8),
          nn.MaxPool2d(kernel_size=2, stride=2),
          
          nn.Conv2d(8, 16, kernel_size = 3, padding = 0, stride = 1),
          nn.ReLU(),
          nn.BatchNorm2d(16),
          nn.MaxPool2d(kernel_size=2, stride=2),


          nn.Conv2d(16, 32, kernel_size = 3, padding = 0, stride = 1),
          nn.ReLU(),
          nn.BatchNorm2d(32),
          nn.MaxPool2d(kernel_size=2, stride=2),

          nn.Conv2d(32, 64, kernel_size = 3, padding = 0, stride = 1),
          nn.ReLU(),
          nn.BatchNorm2d(64),
          nn.MaxPool2d(kernel_size=2, stride=2),

          nn.Conv2d(64, 128, kernel_size = 3, padding = 0, stride = 1),
          nn.ReLU(),
          nn.BatchNorm2d(128),
          nn.MaxPool2d(kernel_size=2, stride=2),          
      )
      self.fc_layers = nn.Sequential(
          nn.Linear(6272, 2048),     
          nn.Linear(2048, 512),
          nn.Linear(512, num_classes)
      )

    def forward(self, x):
      x = self.conv_layers(x)
      x = x.view(x.size(0), -1)
      x = self.fc_layers(x)
      return x
```

## Step 3: Fine-tune a CNN

You will now have to fine-tune a pre-trained CNN on your task. This approach is going to give you much better results than training from scratch, because the pre-trained CNN already has learnt meaningful feature maps that you can re-use. Your fine-tuned CNN should attain at least 50% accuracy on the test set.

Our fine-tuned CNN will be based on the AlexNet model. In practice, we will extract the first layers from AlexNet to compute some feature maps. Then you will have to add additional layers compute the final output.

```python
# Define fine-tuned model
class FineTunedAlexNet(nn.Module):
    
    def __init__(self):
        super().__init__()
        # Load AlexNet model
        alexnet = models.alexnet(pretrained=True)
        # Select feature extraction part
        self.features = alexnet.features
        # Add layers
        self.layers = nn.Sequential(
          nn.Conv2d(256, 256, kernel_size=3, padding=0, stride=1),
          nn.ReLU(),
          nn.BatchNorm2d(256),
          nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.fc_layers = nn.Sequential(
            nn.Linear(1024, 512),
            nn.Linear(512, 256),
            nn.Linear(256, num_classes)
        )

    
    def forward(self, x):
        # Go through the pre-trained layers
        x = self.features(x)
        # Go through added layers
        # HINT: check the output of the model without adding any layer to see the size of the feature maps from AlexNet
        x = self.layers(x)
        x = x.view(x.size(0), -1)
        x = self.fc_layers(x)
        return x
```

```python
# Create model
model = FineTunedAlexNet()
```
